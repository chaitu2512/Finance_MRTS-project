```python
dftop={}
for i in range(0,25):
    for j in range(52):
        c=df_top25.iloc[j].stockname
        d=df_stock.iloc[i][0]
        if(c==d):
            dftop[d]=df_top25.iloc[j]["time_frame"]
            break
dftop
```

```python
df = pd.DataFrame(columns= ["stock_name","TimeFrame"])
```
